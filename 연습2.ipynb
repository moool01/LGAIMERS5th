```python
# 열 이름을 단일 레벨로 변환 ('Set ID'는 두 개의 레벨 중 첫 번째 레벨)
def flatten_columns(df):
    df.columns = ['_'.join(col).strip() for col in df.columns.values]
    return df

auto_clave = flatten_columns(auto_clave)
dam_dispensing = flatten_columns(dam_dispensing)
fill1_dispensing = flatten_columns(fill1_dispensing)
fill2_dispensing = flatten_columns(fill2_dispensing)
```

```python
fill2_dispensing
```

```python
data = auto_clave.merge(dam_dispensing, on='Collect Item:_Set ID', suffixes=('_auto_clave', '_dam_dispensing'))
data = data.merge(fill1_dispensing, on='Collect Item:_Set ID', suffixes=('', '_fill1_dispensing'))
data = data.merge(fill2_dispensing, on='Collect Item:_Set ID', suffixes=('', '_fill2_dispensing'))
```

```python
data
```

```python
# 'Collect Item:_Set ID' 열에서 고유 값 개수 확인
unique_values = data['Collect Item:_Set ID'].unique()
print(f"Unique values count: {len(unique_values)}")

# 중복된 값의 개수 확인
duplicates_count = data['Collect Item:_Set ID'].duplicated().sum()
print(f"Number of duplicated values: {duplicates_count}")
```

```python
duplicate_values = data[data.duplicated(subset=['Collect Item:_Set ID'], keep=False)]
print(duplicate_values[['Collect Item:_Set ID']])
```

```python
# 결측치 처리 (예: 평균값 대체)
data.fillna(data.mean(), inplace=True)
```

```python
# 피쳐와 타겟 분리
X = data.drop(columns=['Set ID', 'target'])
y = data['target']

# 데이터 분할
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=110)
```

```python
# 랜덤 포레스트 모델 학습
model = RandomForestClassifier(random_state=110)
model.fit(X_train, y_train)

# 예측 및 평가
y_pred = model.predict(X_test)
print(f"F1 Score: {f1_score(y_test, y_pred, pos_label='AbNormal')}")

# 피쳐 중요도 시각화
importances = model.feature_importances_
indices = np.argsort(importances)[::-1]

plt.figure(figsize=(12, 8))
plt.title("Feature Importances")
plt.bar(range(X.shape[1]), importances[indices], align="center")
plt.xticks(range(X.shape[1]), X.columns[indices], rotation=90)
plt.xlim([-1, X.shape[1]])
plt.show()
```
